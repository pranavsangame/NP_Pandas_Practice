#### concat along columns

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df3],axis=1)
```

```python
pd.concat([df1,df4],axis=1)
```

## join paramter in concat

```python
pd.concat([df1,df4],axis=1,join='inner')   #keeps the common rows
```

```python
pd.concat([df1,df4],axis=1,join='outer')  #keeps all the rows
```

```python
pd.concat([df1,df3],axis=0,ignore_index=True,join='inner')  # keeps the common columns only
```

```python
pd.concat([df1,df3],axis=0,ignore_index=True,join='outer')  #keeps all columns
```

# Merge

```python
display(df1,df2)
```

```python
merged_df = pd.merge(df1,df2,on='ASCII_Num',how='outer')
merged_df
```

```python
merged_df = pd.merge(df1,df2,on='ASCII_Num',how='inner')
merged_df
```

```python
merged_df = pd.merge(df1,df2,on=['ASCII_Num','char'],how='outer')
merged_df
```

```python
merged_df = pd.merge(df1,df2,on=['ASCII_Num','char'],how='inner')
merged_df
```

```python
merged_df = df1.merge(df2,on=['ASCII_Num','char'],how='outer')
merged_df
```

```python
df4 = pd.DataFrame({'ASCII_Numbers':np.arange(69,76),
                   'characters':[chr(i) for i in range(97,104)]})
df4
```

```python
df1.merge(df4,left_on='ASCII_Num',how='outer')
```

```python
df_z = df1.merge(df4,left_on='ASCII_Num',right_on='ASCII_Numbers',how='outer')
df_z
```

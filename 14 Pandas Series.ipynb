# drop_duplicates()   - for dropping the duplicate values

```python
df = pd.DataFrame({'A':[1,2,1,2],
'B':[100,200,100,500],
'C':[4,4,5,5]})

df
```

```python
df.drop_duplicates(['A'],keep= 'first')
```

```python
df.drop_duplicates(['A'],keep= 'last')
```

```python
df.drop_duplicates(['A','B'],keep= 'first')
```

```python
df.drop_duplicates(['A','B'],keep= 'first')[['A','B']]
```

```python
df = pd.read_csv('movies_data.csv')
df
```

```python
df.drop_duplicates(['genre','content_rating'])[['genre','content_rating']].head(50)
```

```python
df.groupby('genre')['content_rating'].unique().reset_index()
```

```python
df.groupby(['genre'])['content_rating'].agg(['nunique','unique','count']).reset_index()
```

```python
df.groupby(['genre'])['content_rating','star_rating'].agg(['nunique','count']).reset_index()
```

# duplicated in pandas

```python
df = pd.DataFrame({'A':[1,2,1,2],
'B':[100,200,100,500],
'C':[5,4,5,5]})

df
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
df[~(df.duplicated())]
```

```python
df[~(df.duplicated())].reset_index()
```

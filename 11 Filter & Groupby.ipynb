# group by functions

```python
data = {'Name':['Jai','Anuj','Jai','Princi','Gaurav','Anuj','Princi','Abhi','Anuj'],
       'Age':[27,24,22,32,33,29,27,32,32],
       'Address':['Nagpur','Pune','Kanpur','Allahabad','Jodhpur','Kanpur','Allahabad','Aligarh','Pune'],
       'Qualification':['Msc','MA','MCA','PHD','BE','B.com','Msc','MA','B.com'],
       'Score':[24,24,35,45,53,47,50,52,60]}

df = pd.DataFrame(data)
df
```

## groups function in groupby

```python
df.groupby('Name').groups
```

```python
df.groupby('Address').groups
```

## ngroups in groupby

```python
df.groupby('Name').ngroups
```

```python
df.groupby('Address').ngroups
```

## size in groupby

```python
df.groupby('Name').size()
```

## sort in groupby

```python
df.groupby(['Score'],sort=True).mean('Age')  #by default sort is True gives output of grouby in sorted order
```

```python
df.groupby(['Score'],sort=False).mean('Age') # is sort = false then it preserve the actual order of the column on which groupby is applied.
```

## statistical functions on groupby

```python
df.groupby(['Name'],sort=False).mean()
```

```python
df.groupby(['Name'],sort=False).median()
```

```python
df.groupby(['Name'],sort=False)['Age'].median().reset_index()
```

```python
df.groupby(['Name'],sort=False)['Age'].sum().reset_index()
```

```python
df.groupby(['Name'],sort=False)['Age'].min().reset_index()
```

```python
df.groupby(['Name'],sort=False)['Age'].max().reset_index()
```

```python
df.groupby(['Name'],sort=False)['Age'].count().reset_index()
```

## agg functions in groupby

```python
df.groupby(['Name'])['Age'].agg(np.sum)
```

```python
df.groupby(['Name'])['Age'].agg([np.sum,np.mean,np.median,np.std])
```

```python
df.groupby(['Name'])['Address'].agg(pd.Series.mode).reset_index()
```

```python
df.groupby(['Name'],sort=False)['Age'].mean().reset_index()
```

```python
df.groupby(['Name']).agg({'Age':np.mean,'Qualification':'count'})
```

```python
df.groupby(['Name']).agg({'Age':'median','Name':'count'})
```

# Apply function

```python
df
```

```python
def add_2_in_age(variable):
    return variable+2

df['Age']=df['Age'].apply(add_2_in_age)
df
```

```python
def uppercase(var):
    return var.upper()

df['Qualification']=df['Qualification'].apply(uppercase)
df
```

```python
df
```

```python
df['Qualification']=df['Qualification'].apply(lambda x : x.upper())
df
```

```python
df['Qualification']=df['Qualification'].transform(lambda x : x.upper())
df
```

```python
df['Score_updated']=df['Score'].transform(lambda x :((x-x.min()/x.std())*10))
df
```

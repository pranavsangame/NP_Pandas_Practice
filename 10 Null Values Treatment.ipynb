### subset -- dropping null values specific to any column/columns.

```python
df.dropna(subset=['City'])
```

```python
df.dropna(subset=['City','Weight in Kgs'])
```

```python
### how parameter    (all-all mentioned columns should have null values. any-any of the mentioned column should have null values)
```

```python
df.dropna(subset=['City','Weight in Kgs'],how='any',axis=0)
```

```python
df.dropna(subset=['Age in Company','Weight in Kgs'],how='all')
```

```python
df.dropna(subset=[2,5],how='any',axis=1)
```

```python
df.dropna(subset=[1],how='any',axis=1)
```

# Sorting 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Emp_Records.csv')
df
```

### sorting indexes

```python
df.sort_index(axis=0,inplace=True)
df
```

```python
df.sort_index(axis=1)
```

```python
dfx = df.set_index('City')
dfx
```

```python
dfx.sort_index(axis=0)
```

# transpose of a matrix

```python
df = df.T
df
```

```python
df.sort_index(inplace=True)
df.T
```

### sorting values

```python
df = df.T
df
```

```python
df.sort_values('Age in Yrs',ascending=True)
```

```python
df.sort_values('Age in Yrs',ascending=False)
```

```python
df.sort_values(['Weight in Kgs','Salary'],ascending=True)
```

```python
df.sort_values(['Weight in Kgs','Salary'],ascending=[True,False])
```

```python
students = [
           (75, 50, 60, 70),
           (75, 55, 65, 75),
           (75, 35, 45, 25),
           (75, 90, 60, 70),
           (76, 90, 70, 60),
           (90, 80, 70, 60),
           (65, 10, 30, 20)
            ]
  
# Create a DataFrame object from
# list of tuples with columns
# and indices.
details = pd.DataFrame(students, columns =['Hindi', 'Math', 
                                           'Science', 'English'],
                        index = ['Ankit', 'Rahul', 'Aishwarya', 
                                 'Shivangi', 'Priya', 'Swapnil',
                                 'Shaurya'])
  
# sort columns of a Dataframe based 
# on a 'Shivangi' row
display(details)
rslt_df = details.sort_values(by = 'Shivangi', axis = 1)
  
# # show the dataframe
rslt_df
```

#### replacing values in dataframe using loc & iloc function

```python
df.loc[11,['PetalLengthCm']] = 1000
df.loc[[11]]
```

```python
df.iloc[11,3] = 2.6
df.iloc[[11]]
```

```python
df.iloc[:3,1:2] = np.array([100,200,300])
df.iloc[:3,1:2]
df
```

```python
df['SepalLengthCm']=np.random.randint(10,99,150)
df
```

```python
np.random.randint(10,100,[150,2])
```

```python
df.iloc[:3,1:3]=[[1,2],
                 [3,4],
                 [6,7]]
df.iloc[:,1:3]=np.random.randint(10,100,[150,2])
df
```
